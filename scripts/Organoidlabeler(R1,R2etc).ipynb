```python
import pandas as pd

# Read only the header
header = pd.read_csv("2026_UoL_prj_gene_matrix_f.tsv", sep="\t", nrows=0)
all_columns = header.columns.tolist()
sample_columns = all_columns[1:]
```

```python
# Build metadata
metadata = []

for idx, sample_name in enumerate(sample_columns):
    sample_number = idx + 1
    
    if 1 <= sample_number <= 48:
        group = "r1"
    elif 49 <= sample_number <= 96:
        group = "r2"
    elif 350 <= sample_number <= 392:
        group = "r3"
    elif 394 <= sample_number <= 429:
        group = "r4"
    else:
        continue
    
    metadata.append({"sample": sample_name, "group": group})
```

```python
# Save only the assigned samples (prevents non r1-r4 being carried)
meta_df = pd.DataFrame(metadata)
meta_df.to_csv("sample_metadata_filtered.csv", index=False)
```
